```python
class TagDataset:
    def __init__(self, df):
        self.df = df
        #self.df = self.df.drop('OwnerUserId', axis=1)  # Drop the OwnerUserId column
        self.trainMatrix = self.load_rating_file_as_matrix()
        self.testRatings = self.load_rating_file_as_list()
        self.testNegatives = self.load_negative_file()

        self.num_users, self.num_items = self.trainMatrix.shape

    def load_rating_file_as_matrix(self):
        num_users = self.df.shape[0]
        num_items = self.df.shape[1]  # This should give the correct number of columns

        print(f"Number of users: {num_users}, Number of items: {num_items}")  # Debug print

        mat = sp.dok_matrix((num_users, num_items), dtype=np.float32)

        for row in range(num_users):
            for col in range(num_items):
                if self.df.iloc[row, col] > 0:
                    mat[row, col] = self.df.iloc[row, col]

        return mat

    def load_rating_file_as_list(self):
        ratings = []
        for user, row in enumerate(self.df.itertuples(index=False)):
            for item, value in enumerate(row):
                if value > 0:
                    ratings.append([user, item])
        return ratings

    def load_negative_file(self):
        negatives = []
        for row in self.df.itertuples(index=False):
            user_negatives = [item for item, value in enumerate(row) if value == 0]
            negatives.append(user_negatives)
        return negatives
```

## GMF

```python
import numpy as np
import pandas as pd
from keras.optimizers import Adam
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from keras.layers import Input, Embedding, Flatten, Dense, Concatenate, Multiply
from keras.models import Model
from keras.regularizers import l2
```

```python
def get_model(num_users, num_tags, latent_dim, regs=[0,0]):
    user_input = Input(shape=(1,), dtype='int32', name='user_input')
    tag_input = Input(shape=(1,), dtype='int32', name='tag_input')

    user_embedding = Embedding(input_dim=num_users, output_dim=latent_dim, name='user_embedding',
                               embeddings_initializer='normal', embeddings_regularizer=l2(regs[0]), input_length=1)
    tag_embedding = Embedding(input_dim=num_tags, output_dim=latent_dim, name='tag_embedding',
                              embeddings_initializer='normal', embeddings_regularizer=l2(regs[1]), input_length=1)

    user_latent = Flatten()(user_embedding(user_input))
    tag_latent = Flatten()(tag_embedding(tag_input))

    # Use Concatenate and then Multiply to achieve the same effect as merge with mode='mul'
    concat_vector = Concatenate()([user_latent, tag_latent])
    predict_vector = Multiply()([user_latent, tag_latent])

    prediction = Dense(1, activation='sigmoid', kernel_initializer='lecun_uniform', name='prediction')(predict_vector)

    model = Model(inputs=[user_input, tag_input], outputs=prediction)
    return model
```

```python
def make_recommendations(user_index, model, tag_dataset, top_n=10):
    # Check if user index is valid
    if user_index < 0 or user_index >= tag_dataset.num_users:
        raise ValueError(f"User index {user_index} is out of range")

    # Get all tags
    all_tags = list(range(tag_dataset.num_items))

    # Get known tags for the user
    known_tags = tag_dataset.trainMatrix[user_index].nonzero()[1]

    # Create input arrays
    user_input = np.array([user_index] * len(all_tags))
    tag_input = np.array(all_tags)

    # Make predictions
    predictions = model.predict([user_input, tag_input])

    # Create a list of (tag_id, prediction) tuples
    tag_predictions = list(zip(all_tags, predictions.flatten()))

    # Sort by prediction score (highest first)
    tag_predictions.sort(key=lambda x: x[1], reverse=True)

    # Get top N recommendations (excluding known tags)
    recommended_tags = []
    for tag_id, score in tag_predictions:
        if tag_id not in known_tags:
            recommended_tags.append((tag_id, score))
            if len(recommended_tags) == top_n:
                break

    return recommended_tags, known_tags
```

## MLP

```python
import numpy as np
import pandas as pd
import scipy.sparse as sp
from keras.models import Model
from keras.layers import Input, Embedding, Flatten, Dense, Concatenate
from keras.optimizers import Adam
from keras.regularizers import l2
from time import time
import math
import heapq
```

```python
def get_model(num_users, num_items, layers=[64, 32, 16, 8], reg_layers=[0, 0, 0, 0]):
    num_layer = len(layers)
    user_input = Input(shape=(1,), dtype='int32', name='user_input')
    item_input = Input(shape=(1,), dtype='int32', name='item_input')

    MLP_Embedding_User = Embedding(input_dim=num_users, output_dim=layers[0]//2, name='user_embedding',
                                   embeddings_initializer='normal', embeddings_regularizer=l2(reg_layers[0]), input_length=1)
    MLP_Embedding_Item = Embedding(input_dim=num_items, output_dim=layers[0]//2, name='item_embedding',
                                   embeddings_initializer='normal', embeddings_regularizer=l2(reg_layers[0]), input_length=1)

    user_latent = Flatten()(MLP_Embedding_User(user_input))
    item_latent = Flatten()(MLP_Embedding_Item(item_input))

    # Use Concatenate instead of merge
    vector = Concatenate()([user_latent, item_latent])

    for idx in range(1, num_layer):
        layer = Dense(layers[idx], kernel_regularizer=l2(reg_layers[idx]), activation='relu', name=f'layer{idx}')
        vector = layer(vector)

    prediction = Dense(1, activation='sigmoid', kernel_initializer='lecun_uniform', name='prediction')(vector)

    model = Model(inputs=[user_input, item_input], outputs=prediction)
    return model
```

```python
def get_train_instances(train, num_negatives):
    user_input, item_input, labels = [], [], []
    num_users, num_items = train.shape
    for (u, i) in train.keys():
        # Positive instance
        user_input.append(u)
        item_input.append(i)
        labels.append(1)
        # Negative instances
        for t in range(num_negatives):
            j = np.random.randint(num_items)
            while (u, j) in train:
                j = np.random.randint(num_items)
            user_input.append(u)
            item_input.append(j)
            labels.append(0)
    return user_input, item_input, labels
```

```python
def evaluate_model(model, testRatings, testNegatives, K):
    """
    Evaluate the performance (Hit_Ratio, NDCG) of top-K recommendation
    Return: score of each test rating.
    """
    hits, ndcgs = [], []

    print(f"Number of test ratings: {len(testRatings)}")
    print(f"Number of test negatives: {len(testNegatives)}")

    if len(testRatings) == 0 or len(testNegatives) == 0:
        print("Warning: testRatings or testNegatives is empty.")
        return (0, 0)

    for idx in range(min(len(testRatings), len(testNegatives))):
        try:
            (hr, ndcg) = eval_one_rating(model, testRatings[idx], testNegatives[idx], K)
            hits.append(hr)
            ndcgs.append(ndcg)
        except Exception as e:
            print(f"Error evaluating rating {idx}: {str(e)}")

    if len(hits) == 0 or len(ndcgs) == 0:
        print("Warning: No valid evaluations were performed.")
        return (0, 0)

    return (np.array(hits).mean(), np.array(ndcgs).mean())

def eval_one_rating(model, rating, negatives, K):
    items = negatives + [rating[1]]
    users = np.full(len(items), rating[0], dtype='int32')
    predictions = model.predict([users, np.array(items)],
                                batch_size=100, verbose=0)
    map_item_score = {item: pred for item, pred in zip(items, predictions)}
    items.pop()

    # Evaluate top rank list
    ranklist = heapq.nlargest(K, map_item_score, key=map_item_score.get)
    hr = getHitRatio(ranklist, rating[1])
    ndcg = getNDCG(ranklist, rating[1])
    return (hr, ndcg)

def getHitRatio(ranklist, gtItem):
    for item in ranklist:
        if item == gtItem:
            return 1
    return 0

def getNDCG(ranklist, gtItem):
    for i in range(len(ranklist)):
        item = ranklist[i]
        if item == gtItem:
            return math.log(2) / math.log(i+2)
    return 0
```

## NeuMF

```python
import numpy as np
import pandas as pd
import scipy.sparse as sp
import tensorflow as tf
from tensorflow.keras import Model
from tensorflow.keras.layers import Input, Embedding, Flatten, Dense, Multiply, Concatenate, Dropout
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.regularizers import l2
from time import time
```

```python
def get_model(num_users, num_items, mf_dim, layers, reg_layers, reg_mf):
    num_layer = len(layers)

    # Input variables
    user_input = Input(shape=(1,), dtype='int32', name = 'user_input')
    item_input = Input(shape=(1,), dtype='int32', name = 'item_input')

    # Embedding layer
    MF_Embedding_User = Embedding(input_dim = num_users, output_dim = mf_dim, name = 'mf_embedding_user',
                                  embeddings_initializer = 'random_normal', embeddings_regularizer = l2(reg_mf), input_length=1)
    MF_Embedding_Item = Embedding(input_dim = num_items, output_dim = mf_dim, name = 'mf_embedding_item',
                                  embeddings_initializer = 'random_normal', embeddings_regularizer = l2(reg_mf), input_length=1)

    # MLP Embedding
    MLP_Embedding_User = Embedding(input_dim = num_users, output_dim = int(layers[0]/2), name = "mlp_embedding_user",
                                   embeddings_initializer = 'random_normal', embeddings_regularizer = l2(reg_layers[0]), input_length=1)
    MLP_Embedding_Item = Embedding(input_dim = num_items, output_dim = int(layers[0]/2), name = 'mlp_embedding_item',
                                   embeddings_initializer = 'random_normal', embeddings_regularizer = l2(reg_layers[0]), input_length=1)

    # MF part
    mf_user_latent = Flatten()(MF_Embedding_User(user_input))
    mf_item_latent = Flatten()(MF_Embedding_Item(item_input))
    mf_vector = Multiply()([mf_user_latent, mf_item_latent])

    # MLP part
    mlp_user_latent = Flatten()(MLP_Embedding_User(user_input))
    mlp_item_latent = Flatten()(MLP_Embedding_Item(item_input))
    mlp_vector = Concatenate()([mlp_user_latent, mlp_item_latent])

    for idx in range(1, num_layer):
        layer = Dense(layers[idx], activation='relu', name=f'layer{idx}',
                      kernel_regularizer=l2(reg_layers[idx]))
        mlp_vector = layer(mlp_vector)
        mlp_vector = Dropout(0.2)(mlp_vector)  # Add dropout for regularization

    # Concatenate MF and MLP parts
    predict_vector = Concatenate()([mf_vector, mlp_vector])

    # Final prediction layer
    prediction = Dense(1, activation='sigmoid', kernel_initializer='lecun_uniform', name = 'prediction')(predict_vector)

    model = Model(inputs=[user_input, item_input], outputs=prediction)

    return model

def get_recommendations(model, user_id, num_items, dataset, top_n=10):
    # Get the user's existing interactions
    user_interactions = dataset.trainMatrix[user_id].nonzero()[1]

    # Generate predictions for all items for the given user
    user_input = np.full(num_items, user_id)
    item_input = np.arange(num_items)

    predictions = model.predict([user_input, item_input])

    # Create a mask for items the user hasn't interacted with
    mask = np.ones(num_items, dtype=bool)
    mask[user_interactions] = False

    # Apply the mask to the predictions
    masked_predictions = predictions.flatten() * mask

    # Get the indices of the top N predictions
    top_item_indices = masked_predictions.argsort()[-top_n:][::-1]

    # Return the top N recommended item IDs and their scores
    return [(item_id, predictions[item_id][0]) for item_id in top_item_indices]

def main():
    # Hyperparameters
    num_epochs = 100
    batch_size = 256
    mf_dim = 8
    layers = [64, 32, 16, 8]
    reg_mf = 0.01  # Increased regularization
    reg_layers = [0.01, 0.01, 0.01, 0.01]  # Increased regularization
    num_negatives = 4
    learning_rate = 0.001

    # Load data
    dataset = TagDataset(df)
    train, testRatings, testNegatives = dataset.trainMatrix, dataset.testRatings, dataset.testNegatives
    num_users, num_items = dataset.num_users, dataset.num_items

    print(f"Number of users: {num_users}")
    print(f"Number of items: {num_items}")
    print(f"Number of test ratings: {len(testRatings)}")
    print(f"Number of test negatives: {len(testNegatives)}")

    # Build model
    model = get_model(num_users, num_items, mf_dim, layers, reg_layers, reg_mf)
    optimizer = Adam(learning_rate=learning_rate)
    model.compile(optimizer=optimizer, loss='binary_crossentropy')

    # Training model
    for epoch in range(num_epochs):
        t1 = time()
        # Generate training instances
        user_input, item_input, labels = get_train_instances(train, num_negatives, num_items)

        # Training
        hist = model.fit([np.array(user_input), np.array(item_input)],
                         np.array(labels),
                         batch_size=batch_size, epochs=1, verbose=0, shuffle=True)
        t2 = time()

        loss = hist.history['loss'][0]
        print(f'Epoch {epoch+1}/{num_epochs} [{t2-t1:.1f}s]: loss = {loss:.4f}')

    # Save the model
    model.save('NeuMF_model.h5')
    print("Model saved successfully.")

    # Make recommendations for a few users
    for user_id in range(5):  # Get recommendations for the first 5 users
        recommendations = get_recommendations(model, user_id, num_items, dataset, top_n=10)
        print(f"\nTop 10 new recommendations for user {user_id}:")
        for item_id, score in recommendations:
            print(f"Item {item_id}: Score {score:.4f}")

        # Analyze prediction distribution
        all_predictions = model.predict([np.full(num_items, user_id), np.arange(num_items)])
        print(f"Prediction stats for user {user_id}:")
        print(f"  Min: {np.min(all_predictions):.4f}")
        print(f"  Max: {np.max(all_predictions):.4f}")
        print(f"  Mean: {np.mean(all_predictions):.4f}")
        print(f"  Median: {np.median(all_predictions):.4f}")

# Run the main function
if __name__ == '__main__':
    main()
```

## Evaluation

```python
import math
import heapq
import multiprocessing
import numpy as np
from time import time
```

```python
def evaluate_model(model, testRatings, testNegatives, K, num_thread):
    global _model, _testRatings, _testNegatives, _K
    _model = model
    _testRatings = testRatings
    _testNegatives = testNegatives
    _K = K

    hits, ndcgs = [], []
    if num_thread > 1:  # Multi-thread
        pool = multiprocessing.Pool(processes=num_thread)
        res = pool.map(eval_one_rating, range(len(_testRatings)))
        pool.close()
        pool.join()
        hits = [r[0] for r in res]
        ndcgs = [r[1] for r in res]
    else:  # Single thread
        for idx in range(len(_testRatings)):
            (hr, ndcg) = eval_one_rating(idx)
            hits.append(hr)
            ndcgs.append(ndcg)
    return (hits, ndcgs)

def eval_one_rating(idx):
    rating = _testRatings[idx]
    items = _testNegatives[idx]
    u = rating[0]
    gtItem = rating[1]
    items.append(gtItem)
    # Get prediction scores
    map_item_score = {}
    users = np.full(len(items), u, dtype='int32')
    predictions = _model.predict([users, np.array(items)],
                                 batch_size=100, verbose=0)
    for i in range(len(items)):
        item = items[i]
        map_item_score[item] = predictions[i]
    items.pop()

    # Evaluate top rank list
    ranklist = heapq.nlargest(_K, map_item_score, key=map_item_score.get)
    hr = getHitRatio(ranklist, gtItem)
    ndcg = getNDCG(ranklist, gtItem)
    return (hr, ndcg)

def getHitRatio(ranklist, gtItem):
    return 1 if gtItem in ranklist else 0

def getNDCG(ranklist, gtItem):
    for i, item in enumerate(ranklist):
        if item == gtItem:
            return math.log(2) / math.log(i+2)
    return 0
```

```python
import numpy as np
import pandas as pd
from scipy import sparse

def get_top_n_new_tags(user_id, n=5):
    # Create item IDs for this user
    item_ids = np.arange(num_items)

    # Create user IDs array (same user ID repeated for each item)
    user_ids = np.full_like(item_ids, user_id)

    # Make prediction for this user
    user_predictions = model.predict([user_ids, item_ids]).flatten()

    # Get the known tags for this user
    known_tags = set(train[user_id].nonzero()[1])

    # Calculate the sum of each column in the matrix
    column_sums = train.sum(axis=0).A1

    # Check if known_tags is not empty
    if known_tags:
        # Find the minimum column sum for the known tags
        min_known_tag_sum = min(column_sums[list(known_tags)])
        max_known_tag_sum = max(column_sums[list(known_tags)])
        avg = (min_known_tag_sum+max_known_tag_sum)//2
    else:
        # If known_tags is empty, set min_known_tag_sum to a high value or skip this user
        avg = float('inf')
        min_known_tag_sum = float('inf')

    # Create a list of (tag_id, score) tuples for unknown tags
    unknown_tags = [(i, score) for i, score in enumerate(user_predictions)
                    if i not in known_tags and column_sums[i] <= avg]

    # Sort by score in descending order and get top N
    top_n = sorted(unknown_tags, key=lambda x: x[1], reverse=True)[:n]

    return top_n

# Initialize a dictionary to store top N recommendations for each user
top_n_recommendations = {}

# Generate recommendations for each user
for user in range(num_users):
    top_n_recommendations[user] = get_top_n_new_tags(user)

    if user % 100 == 0:
        print(f"Processed user {user}/{num_users}")

# Convert recommendations to a DataFrame
recommendation_df = pd.DataFrame.from_dict(top_n_recommendations, orient='index')
recommendation_df.columns = [f'rec_{i+1}' for i in range(5)]  # Assuming top 5 recommendations
recommendation_df.index.name = 'user_id'

# Save recommendations to CSV
recommendation_df.to_csv('new_tag_recommendations.csv')
print("Recommendations saved to 'new_tag_recommendations.csv'")

# Print example recommendations
print("\nExample recommendations:")
print(recommendation_df.head())
```
